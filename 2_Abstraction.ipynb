#**2.Abstraction** -Showing only what is needed.
Showing only the essential features and hiding the complex internal details

**Example: A smartphone** is a great real life example of data abstraction - you can take pictures and make calls without knowing how signals and storage works

*   User can focus on what the object does instead of how it does it.

---
#Abstract Base class (ABC)
Abstract classes are created sing **ABC module** and **@abstractmethod** decorator.
*   These are classes that cannot be initiated directly. They serve as blueprint for other classes.










```python
from abc import ABC, abstractmethod

class Greet(ABC):
  @abstractmethod
  def say_hii(self):
    pass    # Abstract method

class English(Greet):
  def say_hii(self):
    return "Hello"

e = English()
print(e.say_hii())
```

#Components of Abstraction
Abstraction is made up of key components like **abstract methods, concrete methods, abstract properties** and **class instantiation rules**. These elements works together to definea clear and enforced structure for subclasses while hiding the internal details.


#**1. Abstract Method**

Abstract methods are method declaration without any implementation inside an abstract class.They act as placeholders that forces the subclasses to have their own specific implementation, ensuring consistent structure across derived classes

















```python
from abc import ABC, abstractmethod

class Animal(ABC):
  def make_sound(self):
    pass # Abstract method, no implementation here
```

#**2. Concrete Method**
Fully implemented  methods within an abstract class. Subclasses can inherit and use them directly.Promoting reuse without needing to redefine common functionality










```python
from abc import ABC, abstractmethod

class Animal(ABC):
  @abstractmethod
  def make_sound(self):
    pass  # Abstract class, to be implemented bu subclasses
  def move(self):
    return 'Moving' #Concrete method with implementation
```

#**3. Abstract Properties**
Works like abstract methods but used for properties. These properties are declared with **@property** decorator  and marked as abstract using **@abstractmethod**. subclasses must implement these properties.














```python
from abc import ABC, abstractmethod

class Animal(ABC):
  @property
  @abstractmethod
  def species(self):
    pass #Abstract property, must be implemented by subclasses

class Dog(Animal):
  @property
  def species(self):
    return "Jack"

#Instantiate the concrete subclass
dog = Dog()
print(dog.species)
```

#Abstract clss Instantiation
Abstract classes cannot be instantiated directly.











```python
from abc import ABC, abstractmethod

class Animal(ABC):
  @property
  @abstractmethod
  def species(self):
    pass

animal = Animal() # gives error becasue abstract class is not implemented fully so cannot be instantiated
```
