# Without _ _ str _ _() Method

```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age
h1 = Human("Manoj", 25)
print(h1)
```

# With _ _ str _ _() Method

```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name} ({self.age})"

h1 = Human("Manoj", 25)
print(h1)
```

#Create Methods - just a function inside a class

*   Every method in a class takes **self** as the first parameter(to access that object's data)

* can call them using **object.function_name()**

```python
class human:
  #constructor
  def __init__(self, name, age):
    self.name = name
    self.age = age

  #Greet
  def greet(self):
    print(f"Hii, {self.name}({self.age}), how are you doing today?")

  #Check elegibility
  def age_check(self):
    return self.age >= 18

h1 = human("Manoj", 25)
h2 = human("Zephaniah", 27)

h3 = human("t", 30)

#calling methods
# can call them using   object.function_name()
h1. greet()
print(h1.age_check())

h2.greet()
print(h2.age_check())

h3.greet()
print(h3.age_check())
```

# Class object attribute


```python
class game:
  membership = True #class object attribute
  def __init__(self, name, age):
    if self.membership == True:
      self.name = name
      self.age = age

  def greet(self):
    print(f"Hello, {self.name} ({self.age})")
    return f"Hello, {self.name} ({self.age})"
p1 = game("Manoj", 25)
print(p1.greet())
print(p1.membership)



```

#**@classmethod** - Takes **cls**(class itself) as first parameter

-> where as in default method akes **self** as first parameeter

```python
class student:
  school = "auburn"
  def __init__(self, name, age):
    self.name = name
    self.age = age

  @classmethod
  def add(cls, num1, num2):
    return num1 + num2

print(student.add(2,3))
```

#**@staticmethod** - works like normal function inside a class
*   doesn't take **cls** or **self** as first parameters
*  worrks independently



```python
class math:
  @staticmethod
  def add(num1, num2):
    return num1 + num2

print(math.add(21, 27))
```
