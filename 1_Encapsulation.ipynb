#Access Specifiers  
Define how class members(attributes and methods) can be accessed from outside of class


There are 3 Types of Access Specifiers
1.  Public
2.  Protected ( self._attributeName)
3.  Private ( self._ _ attributeName)






#**1.Public Members**
Are accessible from anywhere inside the class or outside the class or from other module.By default all members in Python are Public

```python
class student:
  def __init__(self, name, course):
    self.name = name     #Public atribute
    self.course = course  #Public atribute

  def welcome(self):
    print(f"Welcome {self.name} into the course {self.course}")

student1 = student("Manoj", "cse")

print(student1.name)   #Accessible

student1.welcome()   #Accessible
```

#**Protected Attributes** ( self._ attributeName)
Accessible within class, in subclasses but not outside of the class



```python
class student:
  def __init__(self, name, course):
    self.name = name     #Public atribute
    self._course = course  #Public atribute

class Cse(student):
  def show_course(self):
    print(f"welcome {self.name} into the course {self._course}")

student1 = Cse("Manoj", 'Cse')
student1.show_course()
print(student1.show_course())  #⚠️ Works in Python, but by convention should not be used directly
```

#**Privvte Attributes**(self._ _ attributeName)

Cannot be accessed from outside of the class

```python
class student:
  def __init__(self, name, course):
    self.name = name     #Public atribute
    self.__course = course  #Public atribute

  def welcome(self):
    print(f"course_name is {self.__course}")

student1 = student("Manoj", "Cs")
print(student1.name)

student1.welcome()

#print(student1.__course) -- Error: Not accessible privately


```

#Bank example

1. balance
2. show balance
3. update balance
4. deposit


```python
class Bank_Account:
  def __init__(self):
    self.balance = 1000

  def _show_balance(self):
    print(f'Available Balance is {self.balance}')

  def __update_balance(self, amount):
    self.balance +=  amount

  def deposit(self, amount):
    if amount  > 0:
      self.__update_balance(amount) # Accessing private method internally
      self._show_balance()  # Accessing protectted method internally

    else:
      print(f"Invalid amount")

account = Bank_Account()
account._show_balance()

#account.__update_balance(500)  - Error:Private method cannot be accessed outisde of the class

account.deposit(500)
```

#**Getter(get_attribute)** and **Setter(set_attribute)**

Are used to access and modify the private attributes safely.Insted of accessing the private data directly these methoda gives controled access
* Read data using a getter method.
*  Update data using a setter method with optional validation or restrictions.


---
Eg; Access the private attribute(__salary) and update it

```python
class Employee:
  def __init__(self):
    self.__salary = 1000

  #Getter method
  def get_salary(self):
    return self.__salary

  def set_salary(self, amount):
    if amount > 0:
      self.__salary = amount

    else:
      print("Invalid amount")

emp= Employee()
print(emp.get_salary())

emp.set_salary(2000)
print(emp.get_salary())
```
