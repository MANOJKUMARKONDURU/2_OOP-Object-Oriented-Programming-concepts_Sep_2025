#**Polymorphism** means **Many Forms**
In programming it refers to methods/functions/operators with the same name that can be executed on many other objects or classes














#Function Polymorphism
An example of a python function that can be used on different objects is the **len()** function


----

----




```python
x = "Hello"
print(len(x))
```

# Class Polymorphism
Polymorphism is often used in Class methods, where we can have multiple classes with same method name.


For example, say we have **three classes: Plane, Car, Boat** and they all have the same method **move()**






```python
class Plane:
  def __init__(self, name, model):
    self.name = name
    self.model = model

  def move(self):
    print("Fly")

class Car:
  def __init__(self, name, model):
    self.name = name
    self.model = model

  def move(self):
    print("Drive")

class Boat:
  def __init__(self, name, model):
    self.name = name
    self.model = model

  def move(self):
    print("Sail")

Plane1 = Plane("Manoj", 2027)
car1 = Car("Lambo", 2027)
Boat1 = Boat("Black Pearl", 2027)

for x in (Plane1, car1, Boat1):
  x.move()
```

#Inheritance Class polymorphism

We can use polymorphism for classes with child classes with same name.


Make a parent class called **Vehichle** and make **Plane, Car, Boat** child classes of **vehicles**, the child classes **inherits** the **vehicle**, but can override them:















```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model
  def move(self):
    print("Sail")

class Plane(Vehicle):
  def move(self):
    print("Fly")

class Car(Vehicle):
  def move(self):
    print("Drive")

class Boat(Vehicle):
  pass


Plane1 = Plane("Manoj", 2027)
Car1 = Car("Lambo", 2027)
Boat1 = Boat("Black Pearl", 2027)

for x in (Plane1, Car1, Boat1):
  print(x.brand)
  print(x.model)
  x.move()
```

In the above example Boat class is empty, but it inherits brand, model and move() from Vehicle

The Plane and Car classes also inherits brand, model and move() from the Vehicle, but they both override the move() method
