#**Inheritance**
Parent/Base Class<<<<--child/Derived class

**Parent class**: is the class that being inherited form, also called as **Base class**
**Child Class** is the class that inherits from another class, also called derived class


---

# create a parent class

```python
# Parent class

class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    return f"{self.name} makes a sound"

dog = Animal("Jack")
print(dog.speak())


dog.speak()
```

# Creating a child class


```python
# Parent class

class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    return f"{self.name} makes a sound"

#child class
class Dog(Animal):
  pass

dog = Dog("Jack")
print(dog.speak())
```

#Use the **super()** function
Python also has **super()** function that will make the child class to inherit all the methods and properties from it's parent
* In **super()** function, you don't need to use the name of the parent element, it will automatically inherit the methods and properties from it's parent.






```python
class Kitten(Animal):
  def __init__(self, name):
    super().__init__(name)

kitten = Kitten("Tiger")
kitten.speak()

```


#Add Properties
add a property called born_year to the kitten










```python
class Kitten(Animal):
  def __init__(self, name):
    super().__init__(name)
    self.born_year = 2007

kitten = Kitten("Tiger")
print(kitten.born_year)
```

To pass 2007 into the Kitten class , add another parameter in the _ _ init _ _() function






```python
class Kitten(Animal):
  def __init__(self, name, born_year):
    super().__init__(name)
    self.born_year = born_year

kitten = Kitten("Tiger", 2007)
print(kitten.born_year)
```
